### Dataset 1: United States Obesity Rates

```python
# Load Dataset

df_1 = gpd.read_file('/content/National_Obesity_By_State.geojson')
df_1
```

```python
# Plot

df_1.plot("Obesity", legend=True)
```

The map displays the states and their obesity rates. Generally, the states towards the southeast have higher obesity percentages, and the states in the northwest have lower obesity percentages.

### Dataset 2: Sanitary Sewer Lines

```python
# Load Dataset

df_2 = gpd.read_file('/content/Sanitary_Availability.geojson')
df_2
```

```python
# Plot

df_2.plot('Available', legend=True)
```

This dataset shows the expected areas where sanitary sewer lines will be extended in Sioux Falls, SD. It seems that sewer lines will only be extended onto the outskirts of the map, as the center of the plot is empty. This is likely due to the already existing lines in the center of the city.

### Dataset 3: Ceded Native American Land

### Dataset 4: Land Tenure Zones

```python
# Load Dataset

df_4 = gpd.read_file('/content/BLM_OR_Land_Tenure_Zones_Polygon.geojson')
df_4
```

```python
df_4.plot('LTZ', legend=True)
```

This is a map of Land Tenure Zones in Washington and Oregon as according to the Bureau of Land Management. The zones seem to change going from east to west, with zone 1 being furthest east and zone 3 furthest west. As expected, Zones 1, 2, and 3 are the largest in comparison to 1a, 1b, 2a, 3a, and UN, which are subzones of the larger zones. As there is a lot of empty space, those areas may not be considered Land Tenure Zones under the BLM.

### Dataset 5: Lilac Bloom Dates

```python
# Load Dataset

df_5 = gpd.read_file('/content/Average_Date_When_Lilacs_Bloom_in_Idaho.geojson')
df_5
```

```python
# Plot

df_5.plot('DATE', legend=True)
```

This map is a plot of the average date that lilacs bloom in Idaho. As there are a lot of days, there are a large number of colors on the map. In general, the center of the state has lilacs blooming in the beginning and end of June, where as the rest of the state varies, either blooming in April or May.
